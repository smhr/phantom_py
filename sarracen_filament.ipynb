```python
import matplotlib.pyplot as plt
import sarracen
```

```python
sdf, sdf_sinks = sarracen.read_phantom('turb28_00064', separate_types='sinks')
```

```python
plt.figure(figsize=(20,5))
ax = sdf.render('rho', x='z', y='x', xlim=(-1.5, 1.5), ylim=(-0.5, 0.5), log_scale=True, x_pixels=1800, y_pixels=1800//3)
ax.scatter(x=sdf_sinks['z'], y=sdf_sinks['x'], color='white', s=1)
plt.savefig('turb28_00064.pdf', dpi=300, bbox_inches="tight")
```

```python
plt.figure(figsize=(8,8))
ax = sdf.render('rho', x='z', y='x', xlim=(-.5, .5), ylim=(-0.5, 0.5), log_scale=True, rotation=[0,0,80])
# ax.scatter(x=sdf_sinks['z'], y=sdf_sinks['x'], color='white', s=1)
plt.savefig('turb28_00064_rotated.pdf', dpi=300, bbox_inches="tight")
```
